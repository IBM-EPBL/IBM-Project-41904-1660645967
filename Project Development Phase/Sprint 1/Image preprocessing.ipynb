```python
import keras
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    rotation_range=180,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(
    rescale=1./255
)
```

```python
x_train = train_datagen.flow_from_directory(r'./content/Forest Fire _ Kaggle1.html',
                                            target_size=(128,128),
                                            batch_size=32,
                                            class_mode='binary'
)
```

```python
x_test = test_datagen.flow_from_directory(r'./content/Forest Fire _ Kaggle1.html',
                                              target_size=(128, 128),
                                              batch_size=32,
                                              class_mode='binary'
)
```
